```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# Some data for plotting
x = [0, 1, 2, 3, 4, 5]
y = [0, 4, 8, 12, 16, 20]
y_err = [0.0, 0.5, 1.0, 1.5, 2.0, 2.5]
x_err = .25

plt.errorbar(x, y, 
             xerr= x_err, yerr=y_err,
             color="red", linestyle= 'dashed',
             marker= 'o', markerfacecolor= "yellow",
             ecolor= "black",capsize=5,
             label='Y valus with erros')
plt.xlabel('x-values')
plt.ylabel('y-values')
plt.title('X vs Y')
plt.grid(alpha=.25)
plt.legend(loc = 'lower right')
plt.show()
```

```python
x = np.arange(0,50,0.1)
y = np.exp(-x / 10) * np.sin(x)

plt.plot(x, y, color='blue', linestyle='solid')
plt.xlabel('x', fontsize = 15, color='darkorange')
plt.ylabel('y', fontsize = 15, color= 'darkorange')
plt.title('e$^-x/10$ & $\sin^2x$')
plt.grid(alpha=.25)
```
